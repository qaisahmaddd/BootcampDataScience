2. Code di bawah menunjukkan nilai min dan max waktu pengambilan data:

```python
df = client.query('''
SELECT MIN(trip_start_timestamp), MAX(trip_start_timestamp)
FROM `ftds-hacktiv8-project.phase0_ftds_026_rmt.taxi_fare`
''').to_dataframe()

df
```

Waktu Min adalah = 2021-01-02 11:00:00+00:00

Waktu Max adalah = 2022-12-31 14:15:00+00:00



```python
# Pembuktian, nilai min dengan menggunakan code sorting by date:
df = client.query('''
SELECT *
FROM `ftds-hacktiv8-project.phase0_ftds_026_rmt.taxi_fare`
ORDER BY trip_start_timestamp
''').to_dataframe()

df
```

```python
# # Pembuktian, nilai max dengan menggunakan code sorting DESC by date:

df = client.query('''
SELECT *
FROM `ftds-hacktiv8-project.phase0_ftds_026_rmt.taxi_fare`
ORDER BY trip_start_timestamp DESC;
''').to_dataframe()

df
```

3.
Code di bawah untuk membandingkan banyaknya perjalanan seluruh perusahaan tahun 2022

```python
df = client.query('''
SELECT trip_start_timestamp,
  EXTRACT(year FROM trip_start_timestamp) AS year
FROM `ftds-hacktiv8-project.phase0_ftds_026_rmt.taxi_fare`
''').to_dataframe()

df
```

```python
df = client.query('''
SELECT company, count(*) AS total_trip_company
FROM `ftds-hacktiv8-project.phase0_ftds_026_rmt.taxi_fare`
WHERE YEAR(trip_start_timestamp) = 2022
GROUP BY company;
''').to_dataframe()

df
```

4. Kode di bawah untuk menunjukkan perusahaan dengan tarif perjalanan per KM termurah:

```python
# Kode ini untuk mencari perusahaan apa yang memiliki fare terendah. Dengan menggunakan group by company dan memilih perusahaan teratas (terendah)

df = client.query('''
SELECT company, MIN(fare) AS cheapest
FROM `ftds-hacktiv8-project.phase0_ftds_026_rmt.taxi_fare`
GROUP BY company
ORDER BY cheapest
LIMIT 1;
''').to_dataframe()

df
```

Synctaxi Inc.	adalah perusahaan dengan fare terendah
