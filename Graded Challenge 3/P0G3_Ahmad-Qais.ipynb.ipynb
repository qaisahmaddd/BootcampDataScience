# WEB SCRAPING

```python
# setup via requests, import BS dan PD
import requests
from bs4 import BeautifulSoup
import pandas as pd
from scipy import stats
from scipy.stats import norm
import numpy as np
```

```python
# Fungsi untuk melakukan scraping pada satu halaman pencarian
def scrape_page(url):
    headers = {'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/71.0.3578.98 Safari/537.36 OPR/58.0.3135.79'}
    response = requests.get(url, headers=headers)

    soup = BeautifulSoup(response.text, 'html.parser')

    products = []

    product_elements = soup.find_all('div', class_='css-llwpbs') # <-- css-llwpbs adalah class produk reguler
    i = 1


    '''
    Setelah mencoba dan mencoba, saya membatasi pencarian 10 produk saja per page
    Hal ini supaya tidaj terlalu banyak nilai kosong
    tetapi pada selanjutnya fungsi tetap dilooping sampai ke page 10
    mendapatkan 90 entries data
    pada tahap ini saya juga berusaha mendapatkan:
    Nama Produk, Harga Produk, Penjual, Kota Toko, Banyaknya Terjual, dan Rating Produk.
    '''
    
    for product in product_elements:
        if i == 10:
            break

        try:
            name = product.find('div', class_='css-3um8ox').text.strip()
            print(name)
        except:
            name = None

        try:
            price = product.find('div', class_='css-h66vau').text.strip()
            print(price)
        except:
            price = None
        
        seller_element = product.find('div', class_='css-1rn0irl')
        try:
            seller = seller_element.find('span', class_='prd_link-shop-name').text.strip()
            print(seller)
        except:
            seller = None

        try:
            city = seller_element.find('span', class_='prd_link-shop-loc').text.strip()
            print(seller)
        except:
            city = None


        rating_sold_element = product.find('div', class_='css-26zmlj')
        try:
            sold = rating_sold_element.find('span', class_='css-1sgek4h').text.strip() if rating_sold_element else None
            print(sold)
        except:
            sold = None

        try:
            rating = rating_sold_element.find('span', class_='css-t70v7i').text.strip() if rating_sold_element else None
            print(rating)
        except:
            rating = None

        products.append({
            'Nama Produk': name,
            'Harga Produk': price,
            'Nama Toko' : seller,
            'Kota Toko': city,
            'Banyaknya Terjual': sold,
            'Rating Produk': rating
        })

        i+=1

    return products
```

```python
# Pada tahap ini proses looping sampai 10 page dilakukan
def scrape_multiple_pages(base_url, num_pages):
    all_products = []

    for page in range(1, num_pages + 1):
        url = f'{base_url}&page={page}'
        products = scrape_page(url)
        all_products.extend(products)

    return all_products

# URL Tokopedia pencarian "seblak"
base_url = 'https://www.tokopedia.com/search?navsource=&page=1&q=seblak&srp_component_id=02.01.00.00&srp_page_id=&srp_page_title=&st='
num_pages = 10  # Jumlah halaman yang ingin di-scrape

# Lakukan scraping
data = scrape_multiple_pages(base_url, num_pages)
```

```python
# Simpan hasil scraping ke dalam pandas DataFrame
df = pd.DataFrame(data)
```

### Potensi Pendapatan Tambahan

```python
# Saya definisikan terlebih dahulu Standard deviasi dan jumlah populasi 
hp_std = df['Harga Produk'].std()
N = len(df)

'''
Dari sini baru bisa saya hitung potensi Pendapatan per Produk dengan
harga Minimum dan harga Maksimum
level confidence = 95%
'''

low_hp, up_hp = stats.norm.interval(0.95,loc=df['Harga Produk'].mean(),scale=hp_std/np.sqrt(N))
print('Potensi Minimum Pendapatan per Produk:',low_hp.round(2))
print('Potensi Maksimum Pendapatan per Produk:',up_hp.round(2))
```

```python
# Saya definisikan terlebih dahulu Standard deviasi dan jumlah populasi 
s_std = df['Banyaknya Terjual'].std()
N = len(df)

'''
Dari sini baru bisa saya hitung potensi Penjualan Produk per Bulan dengan
penjualan Minimum dan Maksimum
level confidence = 95%
'''

low_s, up_s = stats.norm.interval(0.95,loc=df['Banyaknya Terjual'].mean(),scale=s_std/np.sqrt(N))
print('Potensi Minimum Produk Terjual per Bulan:',low_s.round(2))
print('Potensi Maksimum Produk Terjual per Bulan:',up_s.round(2))
```

```python
'''
Saya sudah mendapatkan perhitungan ini:
Potensi Minimum Pendapatan per Produk
Potensi Maksimum Pendapatan per Produk
Potensi Minimum Produk Terjual per Bulan
Potensi Maksimum Produk Terjual per Bulan

Sekarang saatnya mengalikan antara pendapatan per produk dengan penjualan per bulan
'''

min_income = low_hp * low_s
max_income = up_hp * up_s

min_income = min_income.round(2)
max_income = max_income.round(2)

print('Potensi Minimum Pendapatan Tambahan Saya Setiap Bulan:', min_income)
print('Potensi Maksimum Pendapatan Tambahan Saya Setiap Bulan:', max_income)
```

### Perbandingan harga Jabodetabek dan Luar Jabodetabek

Saya harus mengelompokkan terlebih dahulu, kota-kota yang masuk wilayah JABODETABEK
dan luar JABODETABEK

```python
# Sebelumnya saya perlu melihat lagi kota apa saja yang ada di dalam data
list_kota = df['Kota Toko'].unique()
print('kota yang ada di dalm data adalah sebagai berikut:\n',list_kota)
```

```python
# Ok, saya akan mendefinisikan kota mana yang termasuk JABODETABEK
JaBoDeTaBek = ['Jakarta', 'Bogor', 'Depok', 'Tangerang', 'Bekasi']

# Semua kota yang terdapat kata yang ada di JABODETABEK, akan termasuk True. Dan sebaliknya
df['Kawasan'] = df['Kota Toko'].str.contains('|'.join(JaBoDeTaBek), case=False)
df
```

```python
# Terakhir, saya akan ubah nilai True debgab JABODETABEK, dan False Luar JABODETABEK
df['Kawasan'] = df['Kawasan'].map({True: 'JABODETABEK', False: 'Luar JABODETABEK'})
df
```

```python
# Dari sini, pertama saya harus menghitung rata-ratanya
# Menghitung rata-rata harga di kawasan JABODETABEK
avg_jabodetabek = df[df['Kawasan'] == 'JABODETABEK']['Harga Produk'].mean()
avg_jabodetabek = avg_jabodetabek.round()
print('Rata-rata harga per produk di kawasan JABODETABEK adalah: ', avg_jabodetabek)

# Menghitung rata-rata harga di luar kawasan JABODETABEK
avg_luar_jabodetabek = df[df['Kawasan'] == 'Luar JABODETABEK']['Harga Produk'].mean()
avg_luar_jabodetabek = avg_luar_jabodetabek.round()
print('Rata-rata harga per produk di luar kawasan JABODETABEK adalah: ', avg_luar_jabodetabek)
```

Rata-rata harga produk di kawasan JABODETABEK sekitar 1.6 x lebih tinggi daripada rata-rata harga produk di luar kawasan JABODETABEK

H0: Rata-rata harga produk di kawasan JABODETABEK == Luar JABODETABEK
H1: Rata-rata harga produk di kawasan JABODETABEK =! Luar JABODETABEK

```python
hp_JABODETABEK = df[df['Kawasan'] == 'JABODETABEK']['Harga Produk']
hp_luar_JABODETABEK = df[df['Kawasan'] == 'Luar JABODETABEK']['Harga Produk']

t_stat, p_val = stats.ttest_ind (hp_JABODETABEK , hp_luar_JABODETABEK)
print('T-Statistic:',t_stat)
print('P-value:',p_val) 
```

Seperti kita ketahui, semakin besar nilai t-statistic maka semakin besar pula perbedaan antar variabel. Dieperkuat lagi dengan nilai P yang standardnya adalah 0.05, nilai P di sini di bawah nilai tersebut. Artinya, dalam uji statistik ini H0 tidak dapat diterima.

Dengan kata lain, Rata-rata harga produk di JABODETABEK TIDAK SAMA dengan luar Jabodetabek

Dari data rata-rata harga ini saya akan memilih supplier dari Luar JABODETABEK

### Apakah orang lebih cenderung suka dengan produk yang harganya murah?

```python
# Definisikan produk mahal vs murah berdasarkan rata-ratanya
# Jika harga di bawah rata-rata harga, maka murah. Dan sebaliknya

# Ok, saya akan mendefinisikan kota mana yang termasuk JABODETABEK
Kelas_Harga = ['Mahal', 'Murah']

# Semua kota yang terdapat kata yang ada di JABODETABEK, akan termasuk True. Dan sebaliknya
harga_rata_rata= df['Harga Produk'].mean()

df['Kelas Harga'] = df['Harga Produk'].apply(lambda x: 'Murah' if x < harga_rata_rata else 'Mahal')

df
```

```python
# Sekarang kita hitung jumlah penjualan kategori Murah dan Kategori Mahal

terjual_kelas_mahal = df[df['Kelas Harga'] == 'Mahal']['Banyaknya Terjual'].sum()
print('Produk yang terjual dari kelas harga di atas rata-rata adalah:\n' , terjual_kelas_mahal , 'kali')

terjual_kelas_murah = df[df['Kelas Harga'] == 'Murah']['Banyaknya Terjual'].sum()
print('Produk yang terjual dari kelas harga di bawah rata-rata adalah:\n' , terjual_kelas_murah , 'kali')
```

Uji Hipotesa

Kali ini saya akan menggunakan The Mann-Whitney U test 
The Mann-Whitney U test is a nonparametric test (data are not normally distributed). In other words, we use a Mann-Whitney test to determine whether there is a difference in a continuous variable between two independent groups (categorical variable). For example, we want to compare whether there is a difference in IQ score (continuous variable) between male and female students (categorical variable).

reference : https://shorturl.at/uvJY5 | https://shorturl.at/HKR12

H0 : Tidak ada perbedaan signifikan atara kelas harga dengan jumlah produk terjual
H1 : Terdapat perbedaan signifikan atara kelas harga dengan jumlah produk terjual

```python
from scipy.stats import mannwhitneyu

uji_murah = df[df['Kelas Harga'] == 'Murah']['Banyaknya Terjual']
uji_mahal = df[df['Kelas Harga'] == 'Murah']['Banyaknya Terjual']

statistic, p_value = mannwhitneyu(uji_murah, uji_mahal)

print('Nilai U: ', statistic)
print('P-value :', p_value)

alpha = 0.05

if p_value < alpha:
    print('TOLAK H0')
else:
    print('TERIMA H0')
```

Statistik menunjukkan hasil tolak H0
Artinya H1 : Terdapat perbedaan signifikan atara kelas harga dengan jumlah produk terjual
Dapat diterima
