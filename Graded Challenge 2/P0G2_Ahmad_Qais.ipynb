# 2. Data Cleaning

## Data Duplikat

````python
# Cek Duplicate
dataset[dataset.duplicated()]
````

**KNOWLDGE**

Tidak ditemukan adanya data duplikat

## White Space

````python
# Menghapus White Space
dataset.columns = dataset.columns.str.strip()
dataset.info()
````

## Nama Kolom

Setelah step ini, date terlihat sudah lebih rapi. Tapi apakah sudah bersih? Kita akan lakukan tahapan selanjutnya

````python
# Merubah nama kolom menjadi Snake Case
dataset.rename(columns={
    "Segment": "segment",
    "Country": "country",
    "Product": "product",
    "Discount Band": "discount_band",
    "Units Sold": "units_sold",
    "Manufacturing Price": "manufacturing_price",
    "Sale Price": "sale_price",
    "Gross Sales": "gross_sales",
    "Discounts": "discount",
    "Sales": "sales",
    "COGS": "cogs",
    "Profit": "profit",
    "Date": "date"
    }, inplace=True)
````

Ini adalah salah satu poin penting dari nama kolom. Nama kolom tidak mengandung spasi dan kapital. Saya menggunakan *style* ***Snake Case***

## Data Inconsistent

### Ini adalah tahapan terpanjang yaitu membuat data agar menmiliki tipe sebagaimana mestinya

````python
# Menghapus simbol $ dan comma penyebab inconsistent data
dataset['units_sold'] = dataset['units_sold'].str.replace('$', '')
dataset['units_sold'] = dataset['units_sold'].str.replace(',', '')

````

````python
# Merubah tipe data pada kolom Units Sold menjadi Float
dataset['units_sold'] = dataset['units_sold'].astype(float)
dataset.head()
````

````python
# Menghapus simbol $ pada kolom Manufacturing Price penyebab inconsistent data
# Merubah tipe data menjadi Float
dataset['manufacturing_price'] = dataset['manufacturing_price'].str.replace('$', '')
dataset['manufacturing_price'] = dataset['manufacturing_price'].astype(float)
dataset.head()
````

````python
# Menghapus simbol $ pada kolom Sale Price penyebab inconsistent data
# Merubah tipe data menjadi Float
dataset['sale_price'] = dataset['sale_price'].str.replace('$', '')
dataset['sale_price'] = dataset['sale_price'].astype(float)
dataset.head()
````

````python
# Menghapus simbol $ dan comma pada kolom Gross Sales penyebab inconsistent data
# Merubah tipe data menjadi Float

dataset['gross_sales'] = dataset['gross_sales'].str.replace('$', '')
dataset['gross_sales'] = dataset['gross_sales'].str.replace(',', '')
dataset['gross_sales'] = dataset['gross_sales'].astype(float)
dataset.head()
````

````python
# Menghapus simbol $ dan comma pada kolom Sales penyebab inconsistent data
# Merubah tipe data menjadi Float
dataset['sales'] = dataset['sales'].str.replace('$', '')
dataset['sales'] = dataset['sales'].str.replace(',', '')
dataset['sales'] = dataset['sales'].astype(float)
dataset.head()
````

````python
# Menghapus simbol $ dan comma pada kolom COGS penyebab inconsistent data
# Merubah tipe data menjadi Float
dataset['cogs'] = dataset['cogs'].str.replace('$', '')
dataset['cogs'] = dataset['cogs'].str.replace(',', '')
dataset['cogs'] = dataset['cogs'].astype(float)
dataset.head()
````

````python
# kode di bawah ini untuk merubah data supaya bisa dijadikan float
dataset['profit'] = dataset['profit'].str.replace('[\$,()]', '', regex=True)
dataset['profit'] = dataset['profit'].replace('-', '0', regex=True)
dataset['profit'] = dataset['profit'].astype(float)
dataset['profit'] = dataset['profit'].replace('','0').astype(float).fillna(0)
````

````python
# kode di bawah ini untuk merubah data supaya bisa dijadikan float
dataset['discount'] = dataset['discount'].str.replace('[\$,()]', '', regex=True)
dataset['discount'] = dataset['discount'].replace('-', '0', regex=True)
dataset['discount'] = dataset['discount'].astype(float)
dataset['discount'] = dataset['discount'].replace('','0').astype(float).fillna(0)


dataset.info()
````

````python
# Merubah data type pada kolom Date menjadi type DateTime
dataset['date'] = pd.to_datetime(dataset['date'])
````

````python
# Remove 3 kolom ekstraksi waktu
# Menurut saya 3 kolom ini tidak relevan, dan belum dibutuhkan
dataset.drop('Year', axis = 1, inplace=True)
dataset.drop('Month Number', axis = 1, inplace=True)
dataset.drop('Month Name', axis = 1, inplace=True)
````

````python
# Menambahkan kolom transaction_id sebagai primary key tabel_utama
dataset.insert(0, 'transaction_id', range(1, 1 + len(dataset)))
````

````python
# Merubah urutan kolom
# Kolom yang saling terkait didekatkan
# seperti units_sold dan sale_price didekatkan denngan gross_sales
# supaya lebih mudah dibaca

x = pd.DataFrame(dataset)

reorder = ['transaction_id', 'date', 'segment', 'country', 'product', 'discount_band',
          'units_sold', 'sale_price', 'gross_sales', 'discount',
          'sales', 'cogs', 'profit', 'manufacturing_price']

dataset = x.loc[:,reorder]
````

````python
# cek tipe data tiap kolom
# Langkah terkahir dari tahapan ini
dataset.info()
dataset.head()
````

# Data Normalization

````python
tabel_utama = pd.DataFrame(dataset)

tabel_product = pd.DataFrame({
    'product_id': range(1, len(tabel_utama['product'].unique()) + 1),
    'product': tabel_utama['product'].unique()
})

tabel_country = pd.DataFrame({
    'country_id': range(1, len(tabel_utama['country'].unique()) + 1),
    'country': tabel_utama['country'].unique()
})

tabel_segment = pd.DataFrame({
    'segment_id': range(1, len(tabel_utama['segment'].unique()) + 1),
    'segment': tabel_utama['segment'].unique()
})

tabel_discount = pd.DataFrame({
    'discount_id': range(1, len(tabel_utama['discount_band'].unique()) + 1),
    'discount_band': tabel_utama['discount_band'].unique()
})

print(tabel_product,"\n")
print(tabel_country,"\n")
print(tabel_segment,"\n")
print(tabel_discount,"\n")
````

````python
dataset.to_csv('transactions.csv', index=False) # <--- Tabel utama
tabel_product.to_csv('products.csv', index=False)
tabel_country.to_csv('countries.csv', index=False)
tabel_segment.to_csv('segments.csv', index=False)
tabel_discount.to_csv('discounts.csv', index=False)
````

SELECT segment, SUM(profit) AS total_profit_tanpa_diskon
FROM transactions
WHERE discount <> 0
GROUP BY segment;

<!-- Menghitung angka Average, Minimum dan Maximum penjualan berdasarkan negara -->
